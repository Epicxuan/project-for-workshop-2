```python
import requests
import sys
import pandas as pd
from bs4 import BeautifulSoup
import numpy as np
import re,string
import re,string
import csv
punc = '~`!#$%^&*()_+-=|\';":/.,?><~·！@#￥%……&*（）——+-=“：’；、。，？》《{}'
flag=True
for i in range(0,2000):
     #page index
    url=requests.get('http://www.gutenberg.org/ebooks/search/?sort_order=random')
   # print(url)
    html = url.text.encode(url.encoding).decode('utf-8')
    bs = BeautifulSoup(html)
    bookname=[]
    #get book name
    namelist = bs.find_all("span",attrs={"class":"title"})
    for tr in namelist:
        info= list(tr.stripped_strings) 
        bookname.append(info[0]) 
        
    #get author
    bookauthor=[]
    authorlist = bs.find_all("span",attrs={"class":"subtitle"})
    for tr in authorlist:
        info= list(tr.stripped_strings) 
        bookauthor.append(info[0]) 
        
    urllist = bs.find_all("a",attrs={"class":"link"})
    booknum = len(urllist)
    for j in range(3,booknum-1):
         # enter each book website
        rslt=[]    
        name=re.sub(r"[%s]+" %punc, "",bookname[j]) #delete punction
        author= bookauthor[j-3]
        print(name)
        Tem=name.replace(" ", "")
        if Tem.encode('utf-8').isalpha() == False:
            print("Invail book name")
            flag=False
            break
        name=name+'.txt' # bookname
        bookurl=requests.get('http://www.gutenberg.org'+urllist[j].get('href')+'.txt.utf-8')
        
        html = bookurl.text.encode(bookurl.encoding).decode('utf-8')
        bs = BeautifulSoup(html,'lxml')
        article=bs.get_text()
        rslt.append(name)
        rslt.append(author)
        rslt.append('http://www.gutenberg.org'+urllist[j].get('href'))
        print(rslt)
        # creat a new tet file
        try:
            f = open(name,'a',encoding='utf-8')  
            f.write(article) 
            f.close()
        except: print("error")
        try:
            with open("booklist2.csv","a") as csvfile:
                    writer = csv.writer(csvfile)    
                    writer.writerow(rslt)
        except: print("error")
    #print("page :" +str(i)+"done")
print("done")
```
